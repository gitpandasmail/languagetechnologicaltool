```python
import sqlite3
import os

#
#os.chdir("/data/conflict_data/")
#os.chdir("../../")
#os.chdir("loputoo bsc thesis/")
# Get the current working directory
current_directory = os.getcwd()
print("Current Directory:", current_directory)

def create_database(db_file):
  """Creates a new SQLite database file with a sample table.

  Args:
    db_file: The path to the database file.
  """
  try:
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    # Create a table (example: a table for storing books)
    #cursor.execute('''
    #  CREATE TABLE IF NOT EXISTS books (
    #      id INTEGER PRIMARY KEY AUTOINCREMENT,
    #      title TEXT NOT NULL,
    #      author TEXT,
    #      published_year INTEGER
    #  )
    #''')

    # Add some sample data (optional)
    #books_data = [
    #    ('The Hitchhiker\'s Guide to the Galaxy', 'Douglas Adams', 1979),
    #    ('Pride and Prejudice', 'Jane Austen', 1813),
    #    ('To Kill a Mockingbird', 'Harper Lee', 1960)
    #]
    #cursor.executemany("INSERT INTO books (title, author, published_year) VALUES (?, ?, ?)", books_data)


    #conn.commit()  # Save the changes
    #print(f"Database '{db_file}' created successfully.")

  except sqlite3.Error as e:
    print(f"An error occurred: {e}")
  finally:
    if conn:
      conn.close()

def print_all_tables(db_file):
    try:
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()

        cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        tables = cursor.fetchall()

        if tables:
            print("Tables in the database:")
            for table in tables:
                print(table[0])
        else:
            print("No tables found in the database.")

    except sqlite3.Error as e:
        print(f"An error occurred: {e}")
    finally:
        if conn:
            conn.close()


def print_rows(db_file, table_name, columns, where, num_rows=0):
    rows=[]
    if (where!=""):
        where = "WHERE "+where
    try:
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()
        limit = ""
        if (num_rows>0):
            limit = "LIMIT "+str(num_rows)
        query = f"SELECT "+columns+" FROM "+table_name+" "+where+" "+limit
        print(query)
        cursor.execute(query)
        rows = cursor.fetchall()

        if rows:
            print("First "+str(num_rows)+" rows of "+table_name+" table:")
            for row in rows:
                print(row)
        else:
            print("sqlite_sequence table is empty or does not exist.")
    
    except sqlite3.Error as e:
        print(f"An error occurred: {e}")
    
    finally:
        if conn:
            conn.close()
    return rows

def print_table_columns(db_file, table_name):
    try:
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()

        cursor.execute(f"PRAGMA table_info({table_name})")
        columns = cursor.fetchall()

        if columns:
            print(f"Columns in table '{table_name}':")
            for column in columns:
                print(column[1])  # Print the column name (index 1)
        #    print(columns)
        else:
            print(f"Table '{table_name}' not found or has no columns.")

    except sqlite3.Error as e:
        print(f"An error occurred: {e}")
    finally:
        if conn:
            conn.close()
    
# Example usage (assuming the database file is in the same directory)
db_filename = "data/conflict_data/syntax_morphology_conflicts.db" 
#db_filename = "syntax_morphology_conflicts.db" 

print_table_columns(db_filename, "syntax_morphology_conflicts")
#print_all_tables(db_filename)
#rows = print_rows(db_filename,"syntax_morphology_conflicts", "pattern_id,sentence_id,verb_loc,phrase,verb,verb_compound, phrase_root_lemma, current_analysis ,current_case, possible_cases","possible_cases LIKE '%all%'",10000)
#rows1 = print_rows(db_filename,"syntax_morphology_conflicts", "pattern_id,sentence_id,verb_loc,phrase,verb,verb_compound, phrase_root_lemma, current_analysis ,current_case, possible_cases","verb='aasima'",100000)
#rows1 = print_rows(db_filename,"syntax_morphology_conflicts", "pattern_id,sentence_id,verb_loc,phrase,verb,verb_compound, phrase_root_lemma, current_analysis ,current_case, possible_cases","",18000)
#rows1 = print_rows(db_filename,"syntax_morphology_conflicts", "COUNT(*)","",18000)
rows1 = print_rows(db_filename,"syntax_morphology_conflicts", "COUNT(*)","verb='aasima'",100)
print("ridu:")
#print(len(rows1))
print(rows1)
#rows1 = print_rows(db_filename,"syntax_morphology_conflicts", "pattern_id,sentence_id,verb_loc,phrase,verb,verb_compound, phrase_root_lemma, current_analysis ,current_case, possible_cases","verb='aasima'",18000)
#print(rows1)
rows1 = print_rows(db_filename,"syntax_morphology_conflicts", "pattern_id,sentence_id,verb_loc,phrase,verb,verb_compound, phrase_root_lemma, current_analysis ,current_case, possible_cases","",100)

db_filename = "data/filtered_data/vp_data3.db"
#print_all_tables(db_filename)
#print_table_columns(db_filename, "patterns")
#print_rows(db_filename, "patterns", "pattern, verb_word, verb_compound", 50)
#print_rows(db_filename, "patterns", "*", "phrase_case='all'", 50)
rows2 = print_rows(db_filename,"patterns", "COUNT(*)","",18000)
print("ridu:")
#print(len(rows1))
print(rows2)
rows2=print_rows(db_filename, "patterns", "*", "verb_word='aasima'", 50)

db_filename = "data/transactions/v33_koondkorpus_sentences_verb_pattern_obl_20241002-130310.db"
print_all_tables(db_filename)
table="transaction_row"
table="collections_processed"
table="transaction_head"
print_table_columns(db_filename, table)
rows1 = print_rows(db_filename,table, "COUNT(*)","verb='aasima'",18000)
print("ridu:")
#print(len(rows1))
print(rows1)
print_rows(db_filename, table, "*", "verb='aasima'", 5000)

# pat_id, pattern, verb_word, verb_compound, phrase_nr,phrase_case, adp, inf_verb
# Example usage:
 # Replace with your desired filename
create_database(db_filename)
print(len(rows2))
```
